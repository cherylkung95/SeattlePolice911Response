```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
tracts = gpd.read_file("2010_Census_Tracts_for_King_County_-_Conflated_to_Parcels___tracts10_area.geojson")
```

```python
tracts.columns
```

```python
tracts = tracts.drop(['OBJECTID', 'FEATURE_ID', 'TRACT_LBL', 'TRACT_STR',
       'TRACT_FLT', 'TRACT_DEL', 'TRTLABEL_FULL', 'TRTLABEL_COMP',
       'TRTLABEL_TRIM', 'COUNTY_STR', 'COUNTY_INT', 'STATE_STR', 'STATE_INT',
       'LEVEL_1', 'LEVEL_2', 'LEVEL_3', 'TRACT_AREA', 'TRACT_PERI', 'LOGRECNO'], axis =1)
tracts.head()
```

```python
tracts = tracts.rename({'TRACT_INT':'tract'}, axis=1)
```

```python
#tracts['Shape_Area'] = tracts['Shape_Area'].astype(np.float64)
#tracts['AWATER20'] = tracts['AWATER20'].astype(np.int64)
tracts['GEO_ID_TRT'] = tracts['GEO_ID_TRT'].astype(np.int64)
#tracts['ALAND20'] = tracts['ALAND20'].astype(np.int64)
tracts['tract'] = tracts['tract'].astype(np.int64)
```

```python
cluster_tracts = pd.read_csv("Cluster Summaries 2021-12-07\Census Tract_summary_by_cluster.csv")
```

```python
cluster_tracts = cluster_tracts.rename({'Unnamed: 0': 'tract'}, axis=1)
cluster_tracts.head()
```

```python
cluster_tracts = cluster_tracts.fillna(0)
```

```python
cluster_tracts.tract.dtype
```

```python
tracts = tracts.merge(cluster_tracts, how='inner', on='tract')
```

```python
tracts
```

clusters = ['Cluster_1', 'Cluster_2', 'Cluster_3', 'Cluster_4', 'Cluster_5', 'Cluster_6', 'Cluster_7']

for clust in clusters:
    tracts[clust]=tracts[clust].apply(lambda x: x/tracts[clust].sum())

```python
from datetime import date
tracts.to_file(f'tracts_per_clust_{date.today()}.geojson', driver='GeoJSON')
```
